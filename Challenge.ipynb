```python
%%bash
cat > programma.c << 'EOF'

#include <stdio.h>
#include <stdlib.h>

void makeSet(int M, int p[200]);
int Find(int l, int p[200]);
void Union(int m, int n, int p[200]);
double power(int exponent);

int main() {
    FILE *file;
    char x,y[150];
    int i=0,partenza,fine,p[200],row, col, ri, ci, ru, cu;

    file = fopen("labirinto.txt", "r");
    if (file == NULL) {
        printf("Errore nell'apertura del file\n");
        return 1;
    }

    //mettiamo il labirinto in y
    while ((x = fgetc(file)) != EOF) {
        y[i]=x;
        i++;
    }

   //possiamo avere anche labirinti di dimensione superiore all unita
    int dati[100],ind=0,num,b=0;for(int a=0;a<i;a++){
    if(y[a]==' ' || y[a]=='\n'){
        dati[ind]=0;
        while(b>0){
          num=y[a-b]-'0';
          num=num*power(b-1);
          dati[ind]=dati[ind]+num;
          b--;
        }
        ind++;
        b=-1;
      }
      b++;
    }

    //definiamo num righe, num colonne, ingresso e uscita del labirinto
    row = dati[0];
    col = dati[1];
    ri = dati[2];
    ci = dati[3];
    ru = dati[4];
    cu = dati[5];

   int dim=row*col;
   char v[row][col];
    int g=12;
    for(int k=0;k<row;k++){
      for(int l=0;l<col;l++){
        v[k][l]=y[g];
        g++;
      }
      g=g+1;
    }

    makeSet(dim,p);
    partenza=Find((ci-1)+(ri-1)*col, p);

   for(int k=0;k<row;k++){
      for(int l=0;l<col;l++){
        if(v[k][l]!='x'){
        if(l!=col-1){ //possiamo fare il controllo sulla colonna destra
          if(v[k][l]==v[k][l+1] && Find(l+k*col,p)==partenza && Find(l+k*col,p)!=Find(l+1+k*col,p)){ //se il nostro elemento è uguale a quello della colonna destra allora union
            Union(l+k*col, l+1+k*col,p);
          }
        }
        if(k!=row-1){ //possiamo fare il controllo sulla riga inferiore
          if(v[k][l]==v[k+1][l] && Find(l+k*col,p)==partenza && Find(l+k*col,p)!=Find(l+(k+1)*col,p)){ //se il nostro elemento è uguale a quello della riga inferiore allora union
            Union(l+k*col, l+(k+1)*col,p);
          }
        }
        if(l!=0){ //possiamo fare il controllo sulla colonna sinistra
          if(v[k][l]==v[k][l-1] && Find(l+k*col,p)==partenza && Find(l+k*col,p)!=Find(l-1+k*col,p)){ //se il nostro elemento è uguale a quello della colonna sinistra allora union
            Union(l+k*col, l-1+k*col,p);
            l=l-2;
          }
        }
        if(k!=0){ //possiamo fare il controllo sulla riga superiore
          if(v[k][l]==v[k-1][l] && Find(l+k*col,p)==partenza && Find(l+k*col,p)!=Find(l+(k-1)*col,p)){ //se il nostro elemento è uguale a quello della riga superiore allora union
            Union(l+k*col, l+(k-1)*col,p);
            k=k-2;
            l=-1;
          }
        }
        }
      }
    }

    //verifica se esiste un percorso tra partenza e fine
    fine=Find((cu-1)+(ru-1)*col, p);
    if(partenza==fine){
      printf("YES");
    } else {
      printf("NO");
    }

    fclose(file);
    return 0;
}

double power(int exponent){
    double result = 1.0;
    int i;
    if (exponent >= 0){
      for (i=0; i<exponent; i++){
        result *= 10;
      }
    }else{
      for (i=0; i<-exponent; i++){
        result /= 10;
      }
    }
    return result;
}

void makeSet(int M,int p[200]){
      for (int i=0; i<M; i++){
          p[i] = i;
   }
}
int Find(int l, int p[200]) {
  if (p[l] == l)
      return l;
  return Find(p[l],p);
}
void Union(int m, int n, int p[200]) {
  int x = Find(m,p);
  int y = Find(n,p);
  p[y] = x;
}

EOF
gcc programma.c -o programma
./programma
```
